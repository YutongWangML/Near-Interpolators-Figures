```python
from test_error_analytic_form import get_risk_predictors, get_E_tr_inv
from binary_search import binary_search
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 14
import matplotlib.patheffects as pe

from cbcolors import cbcolors
```

```python
import pandas as pd
```

```python
import vlab
config = vlab.utils.load_yaml_file("config.yaml")
```

```python

for i in range(len(config['results'])):
    color = cbcolors[i]
    run_config = vlab.utils.load_yaml_file(f"{config['archive_path']}/{config['results'][i]}/config.yaml")
    results = pd.read_csv(f"{config['archive_path']}/{config['results'][i]}/outputs/results.csv")
    run_config['hidden_layer_sizes']

    medians = results.groupby('tau')[['mlp_E_tr', 'mlp_E_te']].median().reset_index()
    quantiles_20 = results.groupby('tau')[['mlp_E_tr', 'mlp_E_te']].quantile(0.20).reset_index()
    quantiles_80 = results.groupby('tau')[['mlp_E_tr', 'mlp_E_te']].quantile(0.80).reset_index()

    # plt.scatter(results['mlp_E_tr'], results['mlp_E_te'], label=run_config['hidden_layer_sizes'],alpha=0.1,color = color,marker='.')
    label = f"hls = {run_config['hidden_layer_sizes']}"
    plt.plot(medians['mlp_E_tr'], medians['mlp_E_te'], label=label,color =color)
    # plt.plot(quantiles_20['mlp_E_tr'], medians['mlp_E_te'], label=run_config['hidden_layer_sizes'],color = color)
    # plt.plot(quantiles_80['mlp_E_tr'], medians['mlp_E_te'], label=run_config['hidden_layer_sizes'],color = color)
    
    plt.fill_between(quantiles_20['tau'], quantiles_20['mlp_E_te'], quantiles_80['mlp_E_te'], color=color, alpha=0.25, label='20-80 qtl')

    # for _, row in medians.iterrows():
    #     plt.plot(row['mlp_E_tr'], row['mlp_E_te'], 'o',  markersize=10)
# plot_trade_off_curve(gamma=2/3,alpha=1.75, label=r'$\alpha = 1.75$')
# plot_trade_off_curve(gamma=2/3,alpha=1.2, label=r'$\alpha = 1.2$')
plt.axhline(y=1, color='k', linestyle=":", label='noise floor')
plt.legend()
plt.xlabel(r'$\mathcal{E}_{\mathtt{train}}^\ast$')
plt.ylabel(r'$\mathcal{E}_{\mathtt{test}}^\ast$')
plt.title(r'$(\mathcal{E}_{\mathtt{train}}^\ast, \mathcal{E}_{\mathtt{test}}^\ast)$')
plt.savefig('outputs/trade_off.svg')
plt.savefig('outputs/trade_off.png')
```
