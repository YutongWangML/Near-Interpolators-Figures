```python
from test_error_analytic_form import get_risk_predictors, k_grid
from binary_search import binary_search,root_search
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 14
import matplotlib.patheffects as pe

gamma = 0.5
alpha = 1.75
colors = ['#1f77b4', '#419ede']

def get_E_tr_inv(gamma,alpha):
    _, E_tr, R = get_risk_predictors(gamma, alpha)
    k_crit = root_search(R)
    E_tr_inv = lambda tau : binary_search(E_tr, tau, init= k_crit)
    return E_tr_inv
```

```python
E_tr_inv = get_E_tr_inv(gamma,alpha)
ks = np.linspace(k_crit+0.01, k_crit+10, num=100)
plt.plot(ks,E_tr(ks))
plt.axvline(x = k_crit, color='gray', linestyle='--')
plt.axhline(y = E_tr(E_tr_inv(0.1)), color='gray', linestyle='--')
plt.ylabel('E_train')
plt.xlabel('k')
```

```python
k_grid = [E_tr_inv(tau) for tau in np.linspace(0.1,0.9,20)]
```

```python
# Create a new figure and axis or get the current one
fig, ax = plt.subplots()

# Clear the current axis
ax.clear()

ks = np.linspace(k_crit, max(k_grid), num=100)
plt.plot(ks,E_tr(ks))
plt.axvline(x = k_crit, color='gray', linestyle='--')
for k in k_grid:
    plt.axhline(y = E_tr(k), color='gray', linestyle='--')
plt.title("Confirmation that the k_grid is correct")
plt.savefig("outputs/confirm_k_grid.png")
```
