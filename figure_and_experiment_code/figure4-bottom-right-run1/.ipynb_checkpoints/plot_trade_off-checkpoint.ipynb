```python
from vlab._utils import load_yaml_file
import matplotlib.pyplot as plt
config = load_yaml_file('config.yaml')
config
```

```python
import pandas as pd
```

```python
for i in range(len(config['results'])):
    res_hash = config['results'][i]
    archive_path = config['archive_path']

    df = pd.read_csv(f"{archive_path}/{res_hash}/outputs/results.csv")
    run_config = load_yaml_file(f"{archive_path}/{res_hash}/config.yaml")
    
    
    plt.scatter(df['n_tr'], df['mlp_norm'],label=run_config['hidden_layer_sizes'])
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.savefig('outputs/norm_scaling.svg')
plt.savefig('outputs/norm_scaling.png')
```
